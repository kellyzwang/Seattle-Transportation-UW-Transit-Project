```python
month_mapping = {'January': 1, 'February': 2, 'March': 3, 'April': 4,
                 'May': 5, 'June': 6, 'July': 7, 'August': 8,
                 'September': 9, 'October': 10, 'November': 11, 'December': 12}

events['month'] = events['month'].map(month_mapping)

#Combining Month/Day/Year to match other datasets. \
events['date'] = pd.to_datetime(events[['year', 'month', 'day']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d')
```

```python
columns_to_drop = ['year','month','day']
events=events.drop(columns=columns_to_drop)
```

```python
events['day_part_cd']=pd.Series(dtype=str)
```

```python
# Convert 'time' column to datetime.time
events['time'] = pd.to_datetime(events['time'], format='%H:%M').dt.time
events
```

```python

# AM Peak (AM) : 5:00AM - 9:00AM Weekdays
mask_ev1 = ((events['day_of_week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])) 
            & (events['time'] >= pd.to_datetime('5:00:00').time()) 
            & (events['time'] < pd.to_datetime('9:00:00').time()))

# events_copy.loc[mask_ev1, 'AM'] = events_copy['day_part_cd']

# Midday (MID) : 9:00AM - 3:00PM Weekdays
mask_ev2 = ((events['day_of_week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])) 
            & (events['time'] >= pd.to_datetime('9:00:00').time()) 
            & (events['time'] < pd.to_datetime('15:00:00').time()))

#events_copy.loc[mask_ev2, 'MID'] = events_copy['day_part_cd']

# PM PEAK (PM) : 3:00PM - 7:00PM Weekdays
mask_ev3 = ((events['day_of_week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])) 
            & (events['time'] >= pd.to_datetime('15:00:00').time()) 
            & (events['time'] < pd.to_datetime('19:00:00').time()))

#events_copy.loc[mask_ev3, 'PM'] = events_copy['day_part_cd']

# Weekend Day (DAY) : 5:00AM - 7:00PM Weekdays
mask_ev4 = ((events['day_of_week'].isin(['Saturday','Sunday'])) 
            & (events['time'] >= pd.to_datetime('5:00:00').time()) 
            & (events['time'] < pd.to_datetime('19:00:00').time()))

#events_copy.loc[mask_ev4, 'DAY'] = events_copy['day_part_cd']

# Evening (XEV) : 7:00PM - 10:00PM everyday 
mask_ev5 = ((events['day_of_week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])) 
            & (events['time'] >= pd.to_datetime('19:00:00').time()) 
            & (events['time'] < pd.to_datetime('22:00:00').time()))

#events_copy.loc[mask_ev5, 'XEV'] = events_copy['day_part_cd']

# Night (XNT) : 10:00PM - 5:00AM everyday 
mask_ev6 = ((events['day_of_week'].isin(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])) 
            & (events['time'] >= pd.to_datetime('22:00:00').time()) 
            & (events['time'] < pd.to_datetime('5:00:00').time()))

#events_copy.loc[mask_ev6, 'XNT'] = events_copy['day_part_cd']

conditions = [mask_ev1, mask_ev2, mask_ev3, mask_ev4, mask_ev5, mask_ev6]
values = ['AM', 'MID', 'PM', 'DAY', 'XEV', 'XNT']
events['day_part_cd'] = np.select(conditions, values, default='')
```
